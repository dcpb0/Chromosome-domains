```python
import torch
import StructureSampling as sts 
import numpy as np
```

```python
#load interaction matrix
intmat=np.load('UAP_inter_mat_35.npy')
print(f'shape of interaction matrix: {intmat.shape}')
```

```python
#MVN sampling of coordinates

#remove the last row and column of the matrix and unsqueeze to 3D
intmat_m1= torch.from_numpy(intmat[:-1,:-1]).unsqueeze(0) 

#sampling
n_sample=10000
str_sample=sts.generate_coordinates(intmat_m1,num_samples=n_sample)
#set the last monomer as (0,0,0)
array_coordinates=np.concatenate((str_sample,np.zeros((n_sample,3,1))),axis=2)

print(f'shape of coordinate matrix: {array_coordinates.shape}')
```

```python
#Calculate distance
distance_matrix=sts.pairwise_distance_matrix(array_coordinates)
print(f'shape of distance matrix: {distance_matrix.shape}')
```
